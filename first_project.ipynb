```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
# this cause a figure window to open
#%matplotlib notebook 
# this leads to interactive plots embedded within the notebook
sns.set(color_codes=True)

df = pd.read_csv("./Desktop/Juypter lab/data.csv")
df.head(5)
```

```python
df.tail(5)
```

```python
df.dtypes
#checking the types of data
```

```python
df = df.drop(['Engine Fuel Type', 'Market Category', 'Vehicle Style', 'Popularity', 'Number of Doors', 'Vehicle Size'], axis=1)
# Dropping irrelevant columns
df.head()
```

```python
#renaming columns
df = df.rename(columns={"Engine HP":"HP", "Engine Cylinders":"Cylinders", "Transmission Type":"Transmission", "Driven_Wheels":"Drive Mode", "highway MPG":"MPG-H", "city mpg":"MPG-C", "MSRP":"Price"})
df. head()
```

```python
#Dropping the duplicate rows
duplicate_rows_df = df[df.duplicated()]
print("number of duplicate rows:", duplicate_rows_df.shape)
```

```python
#to count the nubmer of rows
df.count()
```

```python
#dropping the duplicate rows
df = df.drop_duplicates()
df.count()
```

```python
#dropping the missing or null values
print(df.isnull().sum())
```

```python
df = df.dropna()
df.count()
```

```python
#box plot is a nice way of detecting the outliers
sns.boxplot(x=df['Price'])
```

```python
#let's plot scatter plot
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(df['HP'], df['Price'])
ax.set_xlabel('HP')
ax.set_ylabel('Price')
plt.show()
```

```python
# to find quntiles and IQR
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
#Now we use IQR to remove the outliers from the data
df = df[~((df < (Q1 - 1.5*IQR)) | (df > (Q3 + 1.5*IQR))).any(axis=1)]
df.shape
```

```python
sns.boxplot(x=df['Price'])
```

```python
#plot heatmaps
plt.figure(figsize=(10,5))
#corr computes pairwise correlation of columns
c = df.corr()
sns.heatmap(c,cmap="BrBG", annot=True)
c
```
